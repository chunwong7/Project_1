```python
#Import modules
import csv
import os
import requests
from pprint import pprint
import pandas as pd
import numpy as np
```

```python
#Fema api only return 1000 rows at once, this particular file has over 167000 rows

url = "https://www.fema.gov/api/open/v1/PublicAssistanceFundedProjectsSummaries?$top=1000"
response = requests.get(url)
data_json = response.json()
#Going through the first 1000 rows
disaster_df1 = pd.DataFrame(data_json['PublicAssistanceFundedProjectsSummaries'])
```

```python
#disaster_df1.head()
#specifying the search: after looking into the data, the data we need are the disaster number,
#state, declaration data, incident type, county, and federal obligated amount
dictionary = {
    'number': [],
    'state': [],
    'county': [],
    'year': [],
    'type': [],
    'amount': []
}

for x in range(len(disaster_df1)):
    dictionary['number'].append(disaster_df1['disasterNumber'][x])
    dictionary['state'].append(disaster_df1['state'][x])
    dictionary['county'].append(disaster_df1['county'][x])
    dictionary['year'].append(disaster_df1['declarationDate'][x].split('-')[0])
    dictionary['type'].append(disaster_df1['incidentType'][x])
    dictionary['amount'].append(disaster_df1['federalObligatedAmount'][x])
```

```python
#retreiving all of the data
#Var skips is used to go through all of the rows 
#because this particular dataset has so much data, we broke the data into two sets

#going through first 85000 rows of data
skips = np.arange(1, 85, 1)
skips = skips * 1000
for x in range(len(skips)):
    url = f"https://www.fema.gov/api/open/v1/PublicAssistanceFundedProjectsSummaries?$top=1000&$skip={skips[x]}"
    response = requests.get(url)
    data_json = response.json()
    disaster_df2 = pd.DataFrame(data_json['PublicAssistanceFundedProjectsSummaries'])
    for y in range(len(disaster_df1)):
        try:
            dictionary['number'].append(disaster_df2['disasterNumber'][y])
        except KeyError:
            dictionary['number'].append('NaN')
        try: 
            dictionary['state'].append(disaster_df2['state'][y])
        except KeyError:
            dictionary['state'].append('NaN')
        try:    
            dictionary['county'].append(disaster_df2['county'][y])
        except KeyError:  
            dictionary['county'].append('NaN')
        try:
            dictionary['year'].append(disaster_df2['declarationDate'][y].split('-')[0])
        except KeyError:
            dictionary['year'].append('NaN')
        try:
            dictionary['type'].append(disaster_df2['incidentType'][y])
        except KeyError:    
            dictionary['type'].append('NaN')
        try:
            dictionary['amount'].append(disaster_df2['federalObligatedAmount'][y])
        except KeyError:
            dictionary['amount'].append('NaN')
    print(url)
```

```python
#going through the rest of the data
skips = np.arange(85, 168, 1)
skips = skips * 1000
for x in range(len(skips)):
    url = f"https://www.fema.gov/api/open/v1/PublicAssistanceFundedProjectsSummaries?$top=1000&$skip={skips[x]}"
    response = requests.get(url)
    data_json = response.json()
    disaster_df2 = pd.DataFrame(data_json['PublicAssistanceFundedProjectsSummaries'])
    for y in range(len(disaster_df2)):
        try:
            dictionary['number'].append(disaster_df2['disasterNumber'][y])
        except KeyError:
            dictionary['number'].append('NaN')
        try: 
            dictionary['state'].append(disaster_df2['state'][y])
        except KeyError:
            dictionary['state'].append('NaN')
        try:    
            dictionary['county'].append(disaster_df2['county'][y])
        except KeyError:  
            dictionary['county'].append('NaN')
        try:
            dictionary['year'].append(disaster_df2['declarationDate'][y].split('-')[0])
        except KeyError:
            dictionary['year'].append('NaN')
        try:
            dictionary['type'].append(disaster_df2['incidentType'][y])
        except KeyError:    
            dictionary['type'].append('NaN')
        try:
            dictionary['amount'].append(disaster_df2['federalObligatedAmount'][y])
        except KeyError:
            dictionary['amount'].append('NaN')
    
    print(url)
```

```python
#converting data into a dataframe
df = pd.DataFrame(dictionary)

#function to remove spaces in the name of states
def remove(string): 
    return string.replace(" ", "")

#creating an array of  names of u.s. territory
states = df['state'].unique()

#headers for the csv
columns = ['disasterNumber', 'county', 'year', 'incidentType', 'amountOblig']

#this loop is going through the array of territories and separate the data and save them
#into individual csv
for x in range(len(states)):
    output_data_file = f"output_data/state_data/{remove(states[x])}Public.csv"
    state_df = df.loc[df['state']==states[x],:]
    state_df = state_df.set_index("state")
    print(output_data_file)
    with open(output_data_file, 'w', newline='') as csvfile:
        csvwriter = csv.writer(csvfile, delimiter=',')
        csvwriter.writerow(columns)
        for y in range(len(state_df)):
            csvwriter.writerow([state_df['number'][y], state_df['county'][y],
                               state_df['year'][y], state_df['type'][y], state_df['amount'][y]])
        
   
```

```python
#create a dictionary for the total amount of money obligated to each state
statetotal = {
    'state': [],
    'totalAmount': []
}

for x in range(len(states)):
    statetotal['state'].append(states[x])
    state = states[x]
    state_csv = df.loc[df['state']==states[x], :]
    state_csv['amount'] = pd.to_numeric(state_csv['amount'])
    statesum = state_csv['amount'].sum()
    statetotal['totalAmount'].append(statesum)
```

```python
#save the state total amount data into a csv
pd.options.display.float_format = '{:.2f}'.format
statecols = ['State', 'Total_Amount']
with open('output_data/state_total.csv', 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile, delimiter=',')
    csvwriter.writerow(statecols)
    for x in range(len(states)):
        csvwriter.writerow([statetotal['state'][x], statetotal['totalAmount'][x]])
statetotal_df = pd.DataFrame(statetotal)
```

```python
#filter the data by counties and saving into csv

columns = ['county', 'totalamount']


for x in range(len(states)):
    state_csv = df.loc[df['state']==states[x], :]
    counties = state_csv['county'].unique()
    output_path = f'output_data/county_data/{remove(states[x])}counties.csv'

    countytotal = {
        'county': [],
        'county_total': []
    
    }
    for y in range(len(counties)):

        county_csv = state_csv.loc[state_csv['county']==counties[y], :]
        countytotal['county'].append(counties[y])
        countyamount = county_csv['amount']
        countytotal['county_total'].append(countyamount.sum())

        
    with open(output_path, 'w', newline='') as csvfile:
        csvwriter = csv.writer(csvfile, delimiter=',')
        csvwriter.writerow(columns)
        for z in range(len(countytotal['county'])):
            csvwriter.writerow([countytotal['county'][z], countytotal['county_total'][z]])
       
    
    
    print(f'success for {states[x]} ')
```
